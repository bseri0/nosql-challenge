### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20

# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint
import pprint

# Connect to the MongoDB server
client = MongoClient('localhost', 27017)

# Select the 'uk_food' database
db = client['uk_food']

# Define the query to find establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to get the number of documents matching the query
count = db.establishments.count_documents(query)

# Display the number of documents
print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
first_document = db.establishments.find_one(query)
pprint.pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame

# Define the query to find establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Retrieve all documents matching the query
all_documents = list(db.establishments.find(query))

# Convert the documents into a Pandas DataFrame
df = pd.DataFrame(all_documents)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

# Use count_documents to display the number of documents in the result

# Display the first document in the results using pprint
query = {
    'LocalAuthorityName': {'$regex': 'London'},
    'RatingValue': {'$gte': 4}  # Assuming RatingValue is stored as an integer or comparable number type
}

# Use count_documents to get the number of documents matching the query
count = db.establishments.count_documents(query)

# Display the number of documents
print(f"Number of establishments in London with a RatingValue >= 4: {count}")

# Display the first document in the results using pprint
first_document = db.establishments.find_one(query)
pprint.pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
# Define the query for establishments in London with a RatingValue >= 4
query = {
    'LocalAuthorityName': {'$regex': 'London'},
    'RatingValue': {'$gte': 4}  # Ensure RatingValue allows for this comparison
}

# Retrieve all documents matching the query
all_documents = list(db.establishments.find(query))

# Convert the documents into a Pandas DataFrame
df = pd.DataFrame(all_documents)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Define the search parameters
degree_search = 0.01
latitude = 51.49042  # Example latitude for "Penang Flavours"
longitude = 0.08384  # Example longitude for "Penang Flavours"

# Define the query
query = {
    'RatingValue': 5,
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}
}

# Define the sort order (ascending order by hygiene score)
sort = [('scores.Hygiene', 1)]

# Define the limit (top 5 results)
limit = 5

# Execute the query, sort, and limit
results = list(db.establishments.find(query).sort(sort).limit(limit))

print(results)
# Print the results
#for result in results:
    #print(result)
```

```python
# Convert result to Pandas DataFrame

# Convert the cursor to a list and then to a DataFrame
df = pd.DataFrame(list(results))

# Display the first few rows of the DataFrame
print(df.head())
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results
# Define the aggregation pipeline
pipeline = [
    {
        '$match': {
            'scores.Hygiene': 0  # Assuming 'Hygiene' score is directly accessible and comparable
        }
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',  # Group by the Local Authority Name
            'count': {'$sum': 1}  # Count the number of documents in each group
        }
    },
    {
        '$sort': {
            'count': -1  # Sort the groups by count in descending order
        }
    }
]

# Execute the aggregation pipeline
results = list(db.establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"Number of documents in the result: {len(results)}")

# Print the first 10 results
for result in results[:10]:
    print(result)
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame

# Convert the aggregation results to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print(df.head(10))
```
